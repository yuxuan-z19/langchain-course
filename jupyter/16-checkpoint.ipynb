```python

import os, uuid, json, re
from typing import TypedDict, Annotated, List

# LangGraph
from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.checkpoint.sqlite import SqliteSaver
from langgraph.prebuilt import ToolNode, tools_condition

# LangChain
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage, ToolMessage
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch

# === 初始化真实 LLM（读取你的环境变量） ===
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL", "gpt-4o-mini"),
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),
    temperature=0.2,
    timeout=1200,
    max_retries=2,
)

# === 新版 Tavily 工具 ===
tavily_tool = TavilySearch(max_results=5)

# 绑定工具，启用模型内生的 tool_calls
llm_with_tools = llm.bind_tools([tavily_tool])

print("✅ 已初始化 ChatOpenAI 与 Tavily（新版）。")
```

```python
class AgentState(TypedDict):
    # 使用 add_messages 自动累计对话消息
    messages: Annotated[List[BaseMessage], add_messages]
    topic: str
    search_results: str
    summary: str
```

```python

def _extract_topic(text: str, fallback: str = "未命名主题") -> str:
    m = re.search(r"'([^']+)'", text)
    if m:
        return m.group(1)
    # 简单回退：裁剪前 80 字
    t = (text or "").strip()
    return t[:80] if t else fallback


def generate_topic_node(state: AgentState):
    """节点1：从用户最后一句话抽取明确主题"""
    print("--- 节点: generate_topic ---")
    sys_prompt = "你是研究助手，请从用户最后一句话中抽取一个具体、明确、可研究的主题，并用单引号括起来返回。只返回主题，不要解释。"
    user_last = state["messages"][-1].content if state.get("messages") else "给我一个研究主题"
    prompt = f"{sys_prompt}\n\n用户最后一句：{user_last}"
    ai = llm.invoke([HumanMessage(content=prompt)])
    topic = _extract_topic(ai.content, "通用主题")
    return {"messages": [ai], "topic": topic}


def llm_decide_node(state: AgentState):
    """节点2：带工具的 LLM 判断是否需要检索，若需要则发起 tool_calls"""
    print("--- 节点: llm_decide ---")
    topic = state.get("topic", "通用主题")
    prompt = (
        f"为主题 '{topic}' 进行搜索，若需要检索请调用 Tavily 工具；"
        "如无需检索，则直接回答“无需检索，直接撰写摘要”。"
    )
    ai = llm_with_tools.invoke([HumanMessage(content=prompt)])
    return {"messages": [ai]}


def route_after_decide(state: AgentState):
    """判断上一条 AI 是否发起了 tool_calls"""
    last = state["messages"][-1]
    return "tools" if getattr(last, "tool_calls", None) else "write_summary"


def collect_search_results_node(state: AgentState):
    """从最新 ToolMessage 收集 Tavily 返回结果，写入 search_results"""
    print("--- 节点: collect_search_results ---")
    for msg in reversed(state["messages"]):
        if isinstance(msg, ToolMessage):
            try:
                content = json.dumps(json.loads(msg.content), ensure_ascii=False, indent=2)
            except Exception:
                content = msg.content
            return {"search_results": content}
    return {"search_results": ""}


def write_summary_node(state: AgentState):
    """节点3：基于（可选）检索结果撰写摘要"""
    print("--- 节点: write_summary ---")
    topic = state.get("topic", "通用主题")
    sr = state.get("search_results", "")
    if sr:
        content = f"基于以下检索结果，为主题 '{topic}' 撰写一段 300 字左右的中文学术性摘要：\n{sr}"
    else:
        content = f"无需检索，请为主题 '{topic}' 直接撰写一段 300 字左右的中文学术性摘要。"
    ai = llm.invoke([HumanMessage(content=content)])
    return {"messages": [ai], "summary": ai.content}
```

```python

builder = StateGraph(AgentState)
builder.add_node("generate_topic", generate_topic_node)
builder.add_node("llm_decide", llm_decide_node)
builder.add_node("tools", ToolNode([tavily_tool]))
builder.add_node("collect_search_results", collect_search_results_node)
builder.add_node("write_summary", write_summary_node)

builder.add_edge(START, "generate_topic")
builder.add_edge("generate_topic", "llm_decide")

builder.add_conditional_edges(
    "llm_decide",
    route_after_decide,
    {"tools": "tools", "write_summary": "write_summary"}
)

builder.add_edge("tools", "collect_search_results")
builder.add_edge("collect_search_results", "write_summary")
builder.add_edge("write_summary", END)

sqlite_path = "checkpoints.sqlite"  # 使用文件持久化，跨单元/重启有效
print("✅ builder 已就绪。稍后进入 with SqliteSaver... 中编译并运行。")
```

```python

from langgraph.checkpoint.sqlite import SqliteSaver

thread_id = str(uuid.uuid4())
config = {"configurable": {"thread_id": thread_id}}
initial_input = {"messages": [HumanMessage(content="我想了解一个关于黑洞的有趣悖论")]}

print(f"🚀 开始新任务：thread_id = {thread_id}\n")

with SqliteSaver.from_conn_string(sqlite_path) as memory:
    graph = builder.compile(checkpointer=memory)

    for event in graph.stream(initial_input, config, stream_mode="values"):
        for key, payload in event.items():
            if key == "__end__":
                vals = payload.get("values", {})
                print("--- 事件: __end__ ---")
                print("最终字段：", {k: vals.get(k) for k in ["topic", "search_results", "summary"]})
                print()
                continue

            if key == "messages":
                print("--- 事件: messages ---")
                try:
                    print(f"新增消息条数：{len(payload)}\n")
                except Exception:
                    print("收到 messages 事件\n")
                continue

            print(f"--- 事件: {key} ---")
            if isinstance(payload, dict):
                print({k: payload.get(k) for k in ["topic", "search_results", "summary"]})
            else:
                print(payload)
            print()

print("🏁 首次运行完成。保存在文件：", sqlite_path)
```

```python

from langgraph.checkpoint.sqlite import SqliteSaver

with SqliteSaver.from_conn_string(sqlite_path) as memory:
    graph = builder.compile(checkpointer=memory)

    history = list(graph.get_state_history(config))

    # 找到：有 topic 且尚未产生 search_results 的检查点
    target = None
    for snap in reversed(history):
        if "topic" in snap.values and "search_results" not in snap.values:
            target = snap
            break

    assert target is not None, "未找到合适检查点，请先成功跑完一次流程。"

    branch_from = target.config
    new_topic = "探索强化学习在机器人控制中的应用"
    branch_cfg = graph.update_state(branch_from, values={"topic": new_topic})

    print("🕰️ 从检查点分叉：", branch_from.get("configurable", {}))
    print("✍️ 改写主题为：", new_topic, "\n")

    for event in graph.stream(None, branch_cfg, stream_mode="values"):
        for key, payload in event.items():
            if key == "__end__":
                vals = payload.get("values", {})
                print("--- 分支事件: __end__ ---")
                print("分支最终：", {k: vals.get(k) for k in ["topic", "search_results", "summary"]})
                print()
                continue

            if key == "messages":
                print("--- 分支事件: messages ---")
                try:
                    print(f"新增消息条数：{len(payload)}\n")
                except Exception:
                    print("收到 messages 事件\n")
                continue

            print(f"--- 分支事件: {key} ---")
            if isinstance(payload, dict):
                print({k: payload.get(k) for k in ["topic", "search_results", "summary"]})
            else:
                print(payload)
            print()

print("✅ 分支执行完成。")
```
