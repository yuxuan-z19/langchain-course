```python
import os

from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_core.messages import HumanMessage

human_msg = HumanMessage(content="你好！我想学习 Python 编程，有什么建议吗？")
print(human_msg.type)  # human
print(human_msg.content)  # 你好！我想学习 Python 编程，有什么建议吗？
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL", "gpt-4o-mini"),
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),  # 若走原生 OpenAI，可传 None/不传
    temperature=0.2,  # 稳定输出
    timeout=30,  # 超时保护（秒）
    max_retries=2,  # 简单重试
)
```

```python
resp = llm.invoke([HumanMessage(content="请介绍下你自己？")])
print(resp.content)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

resp = llm.invoke(
    [
        SystemMessage(content="你是一名脱口秀演员"),
        HumanMessage(content="请介绍下你自己？"),
    ]
)
print(resp.content)
```

```python
import datetime
import json

from langchain_core.messages import FunctionMessage

# 模拟获取当前时间
current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
function_msg = FunctionMessage(
    content=json.dumps(
        {"result": current_time, "timezone": "Asia/Shanghai"}, ensure_ascii=False
    ),
    name="get_current_time",
)

print(function_msg.type)  # function
print(function_msg.name)  # get_current_time
print(function_msg.content)
```
